## Recomendation using collaborative

```python
rating_book
```

```python
x=rating_book.groupby('User-ID').count()['Book-Rating'].reset_index()
x.rename(columns={'Book-Rating':'UserRatingCount'},inplace=True)
x
```

```python
newx=x[x['UserRatingCount']>200]
```

```python
y=newx.merge(rating_book,on='User-ID')
```

```python
yy=y.groupby('Book-Title').count()['Book-Rating'].reset_index()
yy.rename(columns={'Book-Rating':'BookRatingCount'},inplace=True)
```

```python
newy=yy[yy['BookRatingCount']>=50]
```

```python
z=newy.merge(y,on='Book-Title')
```

```python
z
```

```python
pt=z.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sc=cosine_similarity(pt)
```

```python
sc.shape
```

```python
pt.head()
```

```python
def recommend(bookname):
    index=np.where(pt.index==bookname)[0][0]
    similaritems=sorted(list(enumerate(sc[index])),key=lambda x:x[1],reverse=True)[1:6]
    
    for i in similaritems:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```

## Recomendation using collaborative filtering

```python
rating_book
```

```python
x=rating_book.groupby('User-ID').count()['Book-Rating']>200
user200=x[x].index
```

```python
filter200=rating_book[rating_book['User-ID'].isin(user200)]
```

```python
y=filter200.groupby('Book-Title').count()['Book-Rating']>=50
book50=y[y].index
```

```python
filter50=filter200[filter200['Book-Title'].isin(book50)]
```

```python
filter50
```

```python
pt=filter50.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
sc=cosine_similarity(pt)
sc
```

```python
def recommend(bookname):
    index=np.where(pt.index==bookname)[0][0]
    similarity=sorted(list(enumerate(sc[index])),key=lambda x:x[1],reverse=True)[1:6]

    for i in similarity:
        print(pt.index[i[0]])
```

```python
recommend('1984')
```
